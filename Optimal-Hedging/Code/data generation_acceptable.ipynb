```python
class OptionDataCollect:
    def build_synthetic_historical_data(self, merged_data):
        
        all_options = []

       
        expiration_days = np.array([5, 10, 21, 42, 63, 126, 252])

  
        

        for _, row in merged_data.iterrows():
            date = row['date']
            S = float(row['underlying_price'])
            vix_index = float(row['implied_volatility'] * 100)
            rate = float(row['risk_free_rate'])

            tau_array = expiration_days/252

            term_structure = self.vol_term_structure(tau_array, vix_index)

            
    
            for i, tau in enumerate(tau_array):
                
                days = int(expiration_days[i])
                vol_base = float(term_structure.loc[i, 'vol_base']/100)

                a, b, rho, m, sigma_svi = self.SVI_parameters_generation(vol_base, tau, vix_index)

                E = self.strike_grid(S, rate, tau, vix_index, num_strikes = 27)

                k = np.log(E / (S * np.exp(rate * tau)))

                implied_vol_svi = self.SVI_implied_vol(a, b, rho, k, m, sigma_svi, tau)
                   
                   
                d1_array, d2_array = self.d_calculation(S, E, tau, rate, implied_vol_svi)                  
                    
                option_call = self.black_scholes(S, E, tau, rate, d1_array, d2_array, option_type = 'call')
                delta_call = self.delta_calculation(d1_array, option_type = 'call')
                vega_call = self.vega_calculation(S, tau, d1_array)
                        
        
                option_put = self.black_scholes(S, E, tau, rate, d1_array, d2_array, option_type = 'put')
                delta_put = self.delta_calculation(d1_array, option_type = 'put')
                vega_put = self.vega_calculation(S, tau, d1_array)
                 
    
                num_strikes = len(E)  
                
                call_df = pd.DataFrame({
                    'date': [date] * num_strikes,
                    'underlying_price' : [float(S)] * num_strikes,
                    'strike_price' : E.astype(float),
                    'time_to_maturity' : [float(tau)] * num_strikes,
                    'days_to_expire' : [int(days)] * num_strikes,
                    'implied_vol_svi' : implied_vol_svi.astype(float),
                    'risk_free_rate' : [float(rate)] * num_strikes,
                    'option_type' : ['call'] * num_strikes,
                    'option_price' : option_call.astype(float),
                    'delta' : delta_call.astype(float),
                    'vega' : vega_call.astype(float),
                    'd1': d1_array.astype(float)
                })
        
                put_df = pd.DataFrame({
                    'date': [date] * num_strikes,
                    'underlying_price' : [float(S)] * num_strikes,
                    'strike_price' : E.astype(float),
                    'time_to_maturity' : [float(tau)] * num_strikes,
                    'days_to_expire' : [int(days)] * num_strikes,
                    'implied_vol_svi' : implied_vol_svi.astype(float),
                    'risk_free_rate' : [float(rate)] * num_strikes,
                    'option_type' : ['put'] * num_strikes,
                    'option_price' : option_put.astype(float),
                    'delta' : delta_put.astype(float),
                    'vega' : vega_put.astype(float),
                    'd1': d1_array.astype(float)
                })
                
                all_options.append(call_df)
                all_options.append(put_df)

        return pd.concat(all_options, ignore_index = True)

    def strike_grid(self, S, rate, tau, vix_index, num_strikes = 27):

        
        vol = vix_index / 100

        F = S * np.exp(rate * tau)

        stddev = vol * np.sqrt(tau)

         

        

        if vix_index > 35:
            num_stdevs = 2.8
        
        elif vix_index > 25:
            num_stdevs = 2.6

        elif vix_index <12:
            num_stdevs = 2.3

        elif vix_index <18:
            num_stdevs = 2.5
            
        else:
            num_stdevs = 2.6           
        

        min_log_moneyness = -num_stdevs * stddev
        max_log_moneyness = num_stdevs * stddev

        centered_points = np.linspace(-1, 1, num_strikes)

        if tau <= 0.08:
            power = 1.2
        elif tau <= 0.25:
            power = 1.15
        else:
            power = 1.0

        transformed_points = np.sign(centered_points) * np.abs(centered_points)**power
        
        log_moneyness = min_log_moneyness + (transformed_points + 1) * (max_log_moneyness - min_log_moneyness)/ 2
        
       
        strikes = F * np.exp(log_moneyness)

        return strikes

    def vol_term_structure(self, tau, vix_index):

        if not isinstance(tau, (list, np.ndarray)):
            tau = np.array([tau])

        tau_array = np.array(tau)

        low_mid = 1 / (1 + np.exp(-(vix_index - 15) * 0.5))
        mid_high = 1 / (1 + np.exp(-(vix_index - 30) * 0.5))

        tau_array = np.array(tau)
        vol_factor_low = 0.75 + np.log(1 + tau_array) * 0.4
        vol_factor_mid = 0.85 + np.log(1 + tau_array) * 0.3
        vol_factor_high = 1.3 - np.log(1 + tau_array) * 0.3

        factor = (vol_factor_low * (1 - low_mid) + 
                vol_factor_mid * (low_mid - mid_high) +
                vol_factor_high * mid_high)
           
            

        term_vols = vix_index * factor
        
        mask_21day = (tau_array == 21/252)
        if np.any(mask_21day):
            term_vols[mask_21day] = vix_index

             
        
        return pd.DataFrame({'expiry': tau, 'vol_base' : term_vols})
        


    def SVI_parameters_generation(self, vol_base, tau, vix_index):
        
        a = -0.002 * vol_base**2 * tau
        b = 0.22 * vol_base * np.sqrt(tau)
        b = np.clip(b, 0.2, 0.9)

        rho = -0.15 - 0.01 * (vix_index/20 - 1) - 0.01 * np.sqrt(tau)
        rho = np.clip(rho, -0.4, 0.0)

        m = -0.003*(vix_index/20-1) * np.sqrt(tau)

        sigma_svi = 0.25 * np.sqrt(tau) * (1 + 0.4 * vix_index/20)
        sigma_svi = np.clip(sigma_svi, 0.2, 0.8)

        

        return a, b, rho, m, sigma_svi

    def SVI_implied_variance(self, a, b, rho, k, m, sigma_svi):
        
        w = a + b * (rho * (k - m) + np.sqrt((k - m)**2 + sigma_svi**2))

        return w

    def SVI_implied_vol(self, a, b, rho, k, m, sigma_svi, tau):

        w = self.SVI_implied_variance(a, b, rho, k, m, sigma_svi)
        w = np.maximum(w, 1e-10)

        implied_vol_svi = np.sqrt(w/tau)
        implied_vol_svi = np.clip(implied_vol_svi, 0.06, 0.8)

        return implied_vol_svi
        
   


    def d_calculation(self, S, E, tau, r, sigma):

        if np.isscalar(E):
            E = np.array([E])
            sigma = np.array([sigma])

        d1 = (np.log(S/E) + (r + 0.5*sigma**2) * tau ) / (sigma * np.sqrt(tau))
        d2 = d1 - sigma * np.sqrt(tau)
        
        return d1, d2

    def delta_calculation(self, d1, option_type = 'call'):

       
        
        if option_type == 'call':
            delta = norm.cdf(d1)
        else:
            delta = norm.cdf(d1) - 1

        return delta

    def vega_calculation(self, S, tau, d1):

        vega = S * np.sqrt(tau)* norm.pdf(d1)

        return vega
        
    def black_scholes(self, S, E, tau, r, d1, d2, option_type = 'call'):
        
        
        if option_type == 'call':
            option_price = S * norm.cdf(d1) - E * np.exp(-r*tau) * norm.cdf(d2)
    
        else:
            option_price = -S * norm.cdf(-d1) + E * np.exp(-r*tau) * norm.cdf(-d2)

        return option_price
```

```python
def selective_filtering(call_options, put_options):
    
    filter_call = call_options.copy()
    filter_put = put_options.copy()

    mask_call = ~((filter_call['delta'] >= 0.8) &
                    (filter_call['moneyness'] > 1.1))

    mask_call = ~((filter_call['delta'] >= 0.6) &
                    (filter_call['moneyness'] > 1.6))
                        

    mask_call &= ~((filter_call['delta'] <= 0.05) &
                   (filter_call['moneyness'] < 0.8))

   
    
    range_call = (filter_call['moneyness'] > 0.9) & (filter_call['moneyness'] < 1.1)
    delta_range_call = (filter_call['delta'] > 0.3) & (filter_call['delta'] < 0.7 )
    mask_call &= ~(range_call & ~delta_range_call)
  
    

    mask_put = ~((filter_put['delta'] >= -0.2) &
                  (filter_put['moneyness'] > 1.3))

    mask_put &= ~((filter_put['delta'] <=  -0.9) &
                   (filter_put['moneyness'] < 0.8))

   

    range_put = (filter_put['moneyness'] > 0.9) & (filter_put['moneyness'] < 1.1)
    delta_range_put = (filter_put['delta'] > -0.7) & (filter_put['delta'] < -0.4)
    mask_put &= ~(range_put & ~delta_range_put)
    

    filter_call = filter_call[mask_call]
    filter_put = filter_put[mask_put]

    return filter_call, filter_put
    


    
```
