```python
class OptionDataCollect:
    def build_synthetic_historical_data(self, merged_data):
        
        all_options = []

       
        expiration_days = np.array([10, 21, 42, 63, 126, 189, 252])

  
        

        for _, row in merged_data.iterrows():
            date = row['date']
            S = float(row['underlying_price'])
            vix_index = float(row['implied_volatility'] * 100)
            rate = float(row['risk_free_rate']/100)

            tau_array = expiration_days/252

            term_structure = self.vol_term_structure(tau_array, vix_index)

            
    
            for i, tau in enumerate(tau_array):
                
                days = int(expiration_days[i])
                vol_base = float(term_structure.loc[i, 'vol_base']/100)

                a, b, rho, m, sigma_svi = self.SVI_parameters_generation(vol_base, tau, vix_index)

                E = self.strike_grid(S, rate, tau, vix_index, num_strikes = 39)

                k = np.log(E / (S * np.exp(rate * tau)))

                implied_vol_svi = self.SVI_implied_vol(a, b, rho, k, m, sigma_svi, tau)
                   
                   
                d1_array, d2_array = self.d_calculation(S, E, tau, rate, implied_vol_svi)                  
                    
                option_call = self.black_scholes(S, E, tau, rate, d1_array, d2_array, option_type = 'call')
                delta_call = self.delta_calculation(d1_array, option_type = 'call')
                vega_call = self.vega_calculation(S, tau, d1_array)
                        
        
                option_put = self.black_scholes(S, E, tau, rate, d1_array, d2_array, option_type = 'put')
                delta_put = self.delta_calculation(d1_array, option_type = 'put')
                vega_put = self.vega_calculation(S, tau, d1_array)
                 
    
                num_strikes = len(E)  
                
                call_df = pd.DataFrame({
                    'date': [date] * num_strikes,
                    'underlying_price' : [float(S)] * num_strikes,
                    'strike_price' : E.astype(float),
                    'time_to_maturity' : [float(tau)] * num_strikes,
                    'days_to_expire' : [int(days)] * num_strikes,
                    'implied_vol_svi' : implied_vol_svi.astype(float),
                    'risk_free_rate' : [float(rate)] * num_strikes,
                    'option_type' : ['call'] * num_strikes,
                    'option_price' : option_call.astype(float),
                    'delta' : delta_call.astype(float),
                    'vega' : vega_call.astype(float),
                    'd1': d1_array.astype(float)
                })
        
                put_df = pd.DataFrame({
                    'date': [date] * num_strikes,
                    'underlying_price' : [float(S)] * num_strikes,
                    'strike_price' : E.astype(float),
                    'time_to_maturity' : [float(tau)] * num_strikes,
                    'days_to_expire' : [int(days)] * num_strikes,
                    'implied_vol_svi' : implied_vol_svi.astype(float),
                    'risk_free_rate' : [float(rate)] * num_strikes,
                    'option_type' : ['put'] * num_strikes,
                    'option_price' : option_put.astype(float),
                    'delta' : delta_put.astype(float),
                    'vega' : vega_put.astype(float),
                    'd1': d1_array.astype(float)
                })
                
                all_options.append(call_df)
                all_options.append(put_df)

        return pd.concat(all_options, ignore_index = True)

    def strike_grid(self, S, rate, tau, vix_index, num_strikes = 39):

        
        vol = vix_index / 100

        F = S * np.exp(rate * tau)

        stddev = vol * np.sqrt(tau)

         

        

        if vix_index > 35:
            vix_multipler = 1.3
        
        elif vix_index > 25:
            vix_multipler = 1.2

        elif vix_index <12:
            vix_multipler = 0.9

        elif vix_index <18:
            vix_multipler = 1.0
            
        else:
            vix_multipler = 1.1

        
        if tau <= 0.08:
            base_num_stdevs = 1.3
            base_power = 1.0
        elif tau <= 0.25:
            base_num_stdevs = 1.5
            base_power = 1.1
        else:
            base_num_stdevs = 1.7
            base_power = 1.2
        

        num_stdevs = base_num_stdevs * vix_multipler
        power = base_power

        
        
        min_log_moneyness = -num_stdevs * stddev * 1.2
        max_log_moneyness = num_stdevs * stddev * 1.2
        
       
        

        centered_points = np.linspace(-1, 1, num_strikes)


        transformed_points = 0.7 * np.sign(centered_points) * np.abs(centered_points)**power
        
        log_moneyness = min_log_moneyness + (transformed_points + 1) * (max_log_moneyness - min_log_moneyness)/ 2
        
       
        strikes = F * np.exp(log_moneyness)


        strikes = np.sort(strikes)

        return strikes

    def vol_term_structure(self, tau, vix_index):

        if not isinstance(tau, (list, np.ndarray)):
            tau = np.array([tau])

        tau_array = np.array(tau)

        low_mid = 1 / (1 + np.exp(-(vix_index - 18) * 0.3))
        mid_high = 1 / (1 + np.exp(-(vix_index - 32) * 0.3))

        tau_array = np.array(tau)
        vol_factor_low = 0.8 + np.log(1 + tau_array) * 0.3
        vol_factor_mid = 0.9 + np.log(1 + tau_array) * 0.25
        vol_factor_high = 1.2 - np.log(1 + tau_array) * 0.25

        factor = (vol_factor_low * (1 - low_mid) + 
                vol_factor_mid * (low_mid - mid_high) +
                vol_factor_high * mid_high)
           
            

        term_vols = vix_index * factor
        
        mask_21day = (tau_array == 21/252)
        if np.any(mask_21day):
            term_vols[mask_21day] = vix_index

             
        
        return pd.DataFrame({'expiry': tau, 'vol_base' : term_vols})
        


    def SVI_parameters_generation(self, vol_base, tau, vix_index):
        
        a = -0.0008 * vol_base**2 * tau
        b = 0.15 * vol_base * np.sqrt(tau)
        b = np.clip(b, 0.12, 0.65)

        rho = -0.15 - 0.01 * (vix_index/20 - 1) - 0.008 * np.sqrt(tau)
        rho = np.clip(rho, -0.4, 0.0)

        m = -0.002*(vix_index/20-1) * np.sqrt(tau)

        sigma_svi = 0.18 * np.sqrt(tau) * (1 + 0.25 * vix_index/20)
        sigma_svi = np.clip(sigma_svi, 0.15, 0.6)

        

        return a, b, rho, m, sigma_svi

    def SVI_implied_variance(self, a, b, rho, k, m, sigma_svi):
        
        w = a + b * (rho * (k - m) + np.sqrt((k - m)**2 + sigma_svi**2))

        return w

    def SVI_implied_vol(self, a, b, rho, k, m, sigma_svi, tau):

        w = self.SVI_implied_variance(a, b, rho, k, m, sigma_svi)
        w = np.maximum(w, 1e-10)

        implied_vol_svi = np.sqrt(w/tau)
        implied_vol_svi = np.clip(implied_vol_svi, 0.1, 0.6)

        return implied_vol_svi
        
   


    def d_calculation(self, S, E, tau, r, sigma):

        if np.isscalar(E):
            E = np.array([E])
            sigma = np.array([sigma])

        d1 = (np.log(S/E) + (r + 0.5*sigma**2) * tau ) / (sigma * np.sqrt(tau))
        d2 = d1 - sigma * np.sqrt(tau)
        
        return d1, d2

    def delta_calculation(self, d1, option_type = 'call'):

       
        
        if option_type == 'call':
            delta = norm.cdf(d1)
        else:
            delta = norm.cdf(d1) - 1

        return delta

    def vega_calculation(self, S, tau, d1):

        vega = S * np.sqrt(tau)* norm.pdf(d1)

        return vega
        
    def black_scholes(self, S, E, tau, r, d1, d2, option_type = 'call'):
        
        
        if option_type == 'call':
            option_price = S * norm.cdf(d1) - E * np.exp(-r*tau) * norm.cdf(d2)
    
        else:
            option_price = -S * norm.cdf(-d1) + E * np.exp(-r*tau) * norm.cdf(-d2)

        return option_price
```

```python
def minimum_variance(sub_dataframes):
    df = sub_dataframes.copy()
    coeffs_abc = {}
    correction_term_df = pd.DataFrame(columns = [
        'date', 'delta_bucket', 'maturity_bucket','correction_term','window_date'
    ])
    
    for name, sub_df in df.items():
        print(f"Processing subgroup: {name}, initial size: {len(sub_df)}")

      

        sub_df = sub_df.copy()
        
        sub_df['date'] = pd.to_datetime(sub_df['date'])
        sub_df = sub_df.sort_values('date')

        sub_df['option_price_diff'] = sub_df['option_price'] - sub_df['option_price'].shift(1)
        sub_df['underlying_price_diff'] = sub_df['underlying_price'] - sub_df['underlying_price'].shift(1)
        sub_df = sub_df.dropna()
        
        
           

        mask = sub_df['underlying_price_diff'].abs() > 0.05
        sub_df = sub_df[mask].copy()

       
        if len(sub_df) < 21:
            print(f"skipping {name}: insufficient data")
            continue

        window_size = 21
        shifting = 5
        start_date = sub_df['date'].min()
        end_date = sub_df['date'].max()

        a_list = []
        b_list = []
        c_list = []
        date_list = []

        parts = name.split('_')
        delta_key = f"d{parts[0][1:]}_{parts[1]}"
        maturity_key = parts[2]
        

        current_date = start_date
        while current_date +pd.Timedelta(days = window_size-1) <= end_date:
            window_end = current_date + pd.Timedelta(days = window_size-1)
            window_data = sub_df[(sub_df['date'] >= current_date) &
                                (sub_df['date'] <= window_end)].copy()



            if len(window_data) < 10:
                current_date += pd.Timedelta(days = shifting)
                continue
            
        
        
            y = (window_data['option_price_diff']/ window_data['underlying_price_diff']- window_data['delta']) * \
                (window_data['underlying_price'] * np.sqrt(window_data['time_to_maturity']))/ window_data['vega']
            x =  window_data['delta'] 
            
            coeffs = np.polyfit(x, y, 2)
            c, b, a = coeffs[0], coeffs[1], coeffs[2]

            correction_term = (a + b * window_data['delta'] + c * window_data['delta']**2) * \
                window_data['vega']/(window_data['underlying_price'] * np.sqrt(window_data['time_to_maturity']))
            
            a_list.append(a)
            b_list.append(b)
            c_list.append(c)
            date_list.append(current_date)
                                     
            correction_term_rolling = pd.DataFrame({
                'date': window_data['date'],
                'delta_bucket': [delta_key]*len(window_data),
                'maturity_bucket': [maturity_key]*len(window_data),
                'correction_term':correction_term.tolist(),
                'window_date' :[current_date] * len(window_data)
            })
            correction_term_df = pd.concat([correction_term_df, correction_term_rolling], ignore_index = True)                            

            current_date += pd.Timedelta(days = shifting)

        if len(a_list) > 0:
            coeffs_abc[name] = pd.DataFrame({
                'a' : a_list, 
                'b': b_list, 
                'c': c_list
            }, index = date_list)


    return correction_term_df, coeffs_abc
            
```
