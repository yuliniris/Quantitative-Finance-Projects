```python
def create_features(frame):
    df = frame.copy()
     
 
    df['CO'] = df['Close'] / df['Open'] 
    df['HC'] = df['High'] / df['Low']
    df['RET'] = np.log(df['Close'] / df['Close'].shift(1))

    for j in [1,2,3,4,5,6,7]:
        df['Lagret' + str(j)] = df['RET'].shift(j)


    for i in [5, 10, 20, 65]:
        df['PCHG' + str(i)] = df['Close'].pct_change(i)   
        
        df['MA' + str(i)] = df['Close'].rolling(i).mean()
        df['EMA' + str(i)] = df['Close'].ewm(i, adjust = False).mean()
        
        df['RET' + str(i)] = df['RET'].rolling(i).mean()
        df['STD'+ str(i)] = df['RET'].rolling(i).std()


    df['UBol'] = df['Close'].rolling(20).mean() + df['Close'].rolling(20).std() * 2
    df['LBol'] = df['Close'].rolling(20).mean() - df['Close'].rolling(20).std() * 2

    delta = df['Close'].diff()
    gain = (delta.where(delta > 0, 0)).rolling(14).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(14).mean()
    rs = gain / loss
    df['RSI'] = 100 - (100 / (1 + rs))

    df['MACD'] = df['Close'].ewm(12).mean() - df['Close'].ewm(26).mean()
    df['MACD_signal'] = df['MACD'].ewm(9).mean()

    high_low = df['High'] - df['Low']
    high_close = np.abs(df['High'] - df['Close'].shift(1))
    low_close = np.abs(df['Low'] - df['Close'].shift(1))
    
    df['true_range'] = np.maximum(high_low, high_close, low_close)
    df['ATR14'] = df['true_range'].rolling(14).mean()
    
    df['Predict'] = df['Close'].shift(-1)
    df['label'] = np.where(df['Predict'] > 1.0009 *df['Close'], 1, 0)

    df.dropna(inplace=True)

    return df    
```

```python
def model_tuning(hp):
    
    tf.keras.backend.clear_session()   

  
    model = Sequential()
    
 
    hp_units1 = hp.Int('units1', min_value=4, max_value=128, step=4)
    hp_units2 = hp.Int('units2', min_value=4, max_value=64, step=4)
    hp_units3 = hp.Int('units3', min_value=4, max_value=64, step=4)
    hp_units4 = hp.Int('units4', min_value=4, max_value=32, step=4)
   

    
    hp_dropout1 = hp.Float('Dropout_rate', min_value=0, max_value=0.5, step=0.1)
    hp_dropout2 = hp.Float('Dropout_rate', min_value=0, max_value=0.5, step=0.1)
    hp_dropout3 = hp.Float('Dropout_rate', min_value=0, max_value=0.5, step=0.1)
    
     
    hp_learning_rate = hp.Float('learning_rate', min_value=1e-5, max_value=1e-2, sampling='log')
    seqlength = hp.Choice('seqlength', values=[5, 20, 65, 130])
     
    hp_activation1 = hp.Choice(name = 'activation', values = ['relu', 'elu'], ordered = False)
    hp_activation2 = hp.Choice(name = 'activation', values = ['relu', 'elu'], ordered = False)
    hp_activation3 = hp.Choice(name = 'activation', values = ['relu', 'elu'], ordered = False)
    hp_activation4 = hp.Choice(name = 'activation', values = ['relu', 'elu'], ordered = False)
    
    
    model.add(LSTM(hp_units1, 
                   input_shape=(seqlength, featureNo), 
                   activation=hp_activation1, 
                   return_sequences=True, 
                   name='Inputlayer'))   
    model.add(Dropout(hp_dropout1, name='Drouput1'))
    
    model.add(LSTM(hp_units2, 
                   activation = hp_activation2, 
                   return_sequences=True, 
                   name='Hidden01'))
    model.add(Dropout(hp_dropout2, name='Drouput2'))

    model.add(LSTM(hp_units3, 
                   activation = hp_activation3, 
                   return_sequences=True, 
                   name='Hidden02'))
    model.add(Dropout(hp_dropout3, name='Drouput3'))

    model.add(LSTM(hp_units4, 
                   activation = hp_activation4, 
                   return_sequences=False, 
                   name='Hidden03'))
    
    model.add(Dense(units=1, activation='sigmoid', name='Output'))    

    
    opt = Adam(learning_rate=hp_learning_rate, epsilon=1e-08, decay=0.0)    
    #opt = RMSprop(learning_rate=hp_learning_rate, rho=hp.Float('rho', min_value=0.86, max_value=0.99), epsilon=1e-7)
    
     
    model.compile(optimizer=opt, 
                  loss=BinaryCrossentropy(), 
                  metrics=['accuracy', 
                           Precision(),
                           Recall()])

    return model
```

```python
callbackset = [EarlyStopping(patience=10, monitor='val_accuracy', mode='max', verbose=1, restore_best_weights=True),
             TensorBoard(log_dir="./tensorboard/Bytuner_logs")]
```

```python
tuner = BayesianOptimization(
                                model_tuning,
                                objective="val_accuracy",
                                max_trials=12,
                                executions_per_trial=3,
                                num_initial_points=2,
                                hyperparameters=None,
                                tune_new_entries=True,
                                allow_new_entries=True,
                                overwrite=True,
                                directory="./keras",
                                project_name="Bytrial")
```

```python
tuner.search(g, 
               epochs=60, 
               validation_data=g_, 
               callbacks=callbackset, 
               #class_weight=class_weight,
               shuffle=False)
```

```python
best_hp_By = tuner.get_best_hyperparameters()[0]
print(best_hp_By.values)
```

```python
best_model = tuner.hypermodel.build(best_hp_By)
```

```python
history = best_model.fit(
    g,
    epochs=60,
    validation_data=g_,
    callbacks=callbackset,
    #class_weight=class_weight,
    shuffle=False
)
```
