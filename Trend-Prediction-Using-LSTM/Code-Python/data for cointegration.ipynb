```python
fig, ax1 = plt.subplots(figsize=(16, 9))
ax1.plot(dfsp500.index, dfsp500['Close_SPY'], label='Close_SPY')
ax1.plot(dfQQQ.index, dfQQQ['Close_QQQ'], label='Close_QQQ')
ax1.plot(dfDIA.index, dfDIA['Close_DIA'], label='Close_DIA')
ax1.set_ylim(0, 620)
ax1.set_ylabel('SPY QQQ DIA(0-620)', color='black')  
ax1.tick_params(axis='y', labelcolor='black')

ax2 = ax1.twinx()
ax2.plot(dfVXX.index, dfVXX['Close_VXX'], label='Close_VXX')
ax2.set_ylim(0, 4800)   
ax2.set_ylabel('VXX (0-4800)', color='black')   
ax2.tick_params(axis='y', labelcolor='black') 

ax1.legend(loc='upper left')
ax2.legend(loc='upper right') 

plt.show()
```
